```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Read dataset using Pandas
df = pd.read_csv('https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/titanic.csv')
```

```python
df.head()
```

```python
X = pd.get_dummies(df.drop(['PassengerId', 'Survived'], axis=1))
y = df['Survived']
# y = df['Survived'].apply(lambda x: 1 if x=='Yes' else 0) # encode yes-no values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=2)
```

```python
X_train.head()
```

```python
y_train.head()
```

## Import dependencies

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score
```

## Build & Compile Model

```python
model = Sequential()
model.add(Dense(32, activation='relu', input_dim=len(X_train.columns))) # input layer
model.add(Dense(64, activation='relu')) # Hidden Layes
model.add(Dense(1, activation='sigmoid')) # Output Layer
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

## Fit, Predict and Evaluate

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
# Prediction
y_hat = model.predict(X_test)
y_hat = [0 if val<0.5 else 1 for val in y_hat]
y_hat
```

```python
accuracy_score(y_test, y_hat)
```
