```python
# LSTM needs to provide data in 3D format - reshape the data
#Process the data for LSTM
trainX =np.array(X_train)
testX =np.array(X_test)
X_train = trainX.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = testX.reshape(X_test.shape[0], 1, X_test.shape[1])
```

```python
#Building the LSTM Model
lstm_model = Sequential()
lstm_model.add(LSTM(32, input_shape=(1, trainX.shape[1]), activation='relu', return_sequences=False))
lstm_model.add(Dense(1))
```

```python
# Compile the LSTM Model
lstm_model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
plot_model(lstm_model, show_shapes=True, show_layer_names=True)
```

```python
#Model Training
history=lstm_model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=1, shuffle=False)
```

```python
def plot_loss_acc(history):
  '''Plots the training and validation loss and accuracy from a history object'''
  acc = history.history['accuracy']
  #val_acc = history.history['val_accuracy']
  loss = history.history['loss']
  #val_loss = history.history['val_loss']

  epochs = range(len(acc))

  plt.plot(epochs, acc, 'bo', label='Training accuracy')
  #plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
  #plt.title('Training and validation accuracy')
  plt.title('Training accuracy')

  plt.figure()

  plt.plot(epochs, loss, 'bo', label='Training Loss')
  #plt.plot(epochs, val_loss, 'b', label='Validation Loss')
  #plt.title('Training and validation loss')
  plt.title('Training loss')
  plt.legend()

  plt.show()
```

```python
# Plot training results
plot_loss_acc(history)
```
